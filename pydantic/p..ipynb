```python
import os
from dotenv import load_dotenv
from langchain_groq import ChatGroq

load_dotenv()

model = ChatGroq(
    model="llama-3.3-70b-versatile",
    api_key=os.getenv("GROQ_API_KEY")
)
```

PYDANTIC

```python
from pydantic import BaseModel, Field
class movie(BaseModel):
    title: str = Field(description="The title of the movie")
    director: str = Field(description="The director of the movie")
    release_year: int = Field(description="The year the movie was released")
    rating:float = Field(description= " The rating of the movie on a scale of 1 to 10")
```

```python
# this will give us the structured output in the form of a pydantic model and the ai message content too
model_structure=model.with_structured_output(movie,include_raw=True)
model_structure.invoke("i wannt the details of the movie Inception")
```

```python
model_structure=model.with_structured_output(movie)
model.invoke("i wannt the details of the movie Inception")
```

```python
from langchain.agents import create_agent
import os
from dotenv import load_dotenv
from langchain_groq import ChatGroq

load_dotenv()

model = ChatGroq(
    model="llama-3.3-70b-versatile",
    api_key=os.getenv("GROQ_API_KEY")
)

class contactinfo(BaseModel):
    email:str = Field(description="The email address of the contact")
    phone:str = Field(description="The phone number of the contact")
    name:str = Field(description="The name of the contact")
    
agent_contact=create_agent(
    model=model,
    response_format=contactinfo
)

result=agent_contact.invoke({"messages":[{"role":"user","content":"extract the contact information from the following text: 'John Doe can be reached at john.doe@example.com or (555) 123-4567'"}]})
result["structured_response"]
```

```python
### data class

from langchain.agents import create_agent
import os
from dotenv import load_dotenv
from langchain_groq import ChatGroq
from dataclasses import dataclass

load_dotenv()

model = ChatGroq(
    model="llama-3.3-70b-versatile",
    api_key=os.getenv("GROQ_API_KEY")
)

@dataclass
class contactinfo:
    email: str
    phone: str
    name: str

agent_contact=create_agent(
    model=model,
    response_format=contactinfo
)

result=agent_contact.invoke({"messages":[{"role":"user","content":"extract the contact information from the following text: 'John Doe can be reached at john.doe@example.com or (555) 123-4567'"}]})
result["structured_response"]
```
