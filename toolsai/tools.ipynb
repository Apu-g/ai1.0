```python
import os
from dotenv import load_dotenv
from langchain_groq import ChatGroq

load_dotenv()

model = ChatGroq(
    model="qwen/qwen3-32b",
    api_key=os.getenv("GROQ_API_KEY")
)

response = model.invoke("how to parrots fly")

print(response.content)
```

```python
from langchain.tools import tool

@tool
def get_weather(location: str) -> str:
    """get weather information for a given location"""
    # In a real implementation, this would call a weather API.
    return f"The current weather in {location} is sunny with a high of 25°C."

modeltool = model.bind_tools([get_weather])

response = modeltool.invoke("what is the weather like in New York?")
print(response)
```

```python
messages = [
    {
        "role": "user",
        "content": "What's the weather in Boston?"
    }
]

# Model generates tool calls
ai_msg = modeltool.invoke(messages)

# Add AI response
messages.append(ai_msg)

print (messages)

# Step 2: Execute tool calls
for tool_call in ai_msg.tool_calls:

    # Execute tool
    tool_result = get_weather.invoke(tool_call)

    # Add tool result to messages
    messages.append(tool_result)

print(messages)

# Step 3: Send tool results back to model
final_response = modeltool.invoke(messages)

# Print final answer
print(final_response.content)
```
