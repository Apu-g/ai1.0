```python
import os
from dotenv import load_dotenv
from langchain_groq import ChatGroq

load_dotenv()

model = ChatGroq(
    model="llama-3.3-70b-versatile",
    api_key=os.getenv("GROQ_API_KEY")
)

for chunk in model.stream("write a 300 to 500 word essay on the importance of AI in modern society"):
    print(chunk.text, end="")
```

```python
for chunk in model.stream("why do parrots have so colourful feathers?"):
    print(chunk.text, end="", flush=True)
```

```python
responses = model.batch([
    "What is AI in few words?",
    "What is Docker in 50 words?",
    "What is Kubernetes in 100 words?"
])

for r in responses:
    print(r.content)
```
